```python
documents = [
{'type': 'passport', 'number': '2207 876234', 'name': 'Василий Гупкин'},
{'type': 'invoice', 'number': '11-2', 'name': 'Геннадий Покемонов'},
{'type': 'insurance', 'number': '10006', 'name': 'Аристарх Павлов'}
]
directories = {
'1': ['2207 876234', '11-2'],
'2': ['10006'],
'3': []
}

def p (num):
    i=0
    while i<len(documents) and documents[i]['number']!=num :
        i+=1
    return i


def s(num):
    for i in directories:
        if directories[i].count(num)!=0:
            return int(i)
    return (int(i)+1)


def l():
    for i in documents:
        print('№: ',i['number'],' тип: ',i['type'],' владелец: ',i['name'],'Документ хранится на полке: ',s(i['number']))
    return 


def a_s(num):
    if directories.get(num) is None:
        directories[num]=[]        
        print('Полка добавлена. Текущий перечень полок: ',*list(directories.keys()), sep='  ' )
    else:
        print('Такая полка уже существует. Текущий перечень полок: ',*list(directories.keys()), sep='  ')
        
        
def ds(num):
    if directories.get(num) is None:
        print('Такой полки не существует. Текущий перечень полок: ',*list(directories.keys()), sep='  ')
    elif directories[num]!=[]:
            print('На полке есть документа, удалите их перед удалением полки. Текущий перечень полок: ',*list(directories.keys()), sep='  ')
    else: 
            del(directories[num])
            print('Полка удалена. Текущий перечень полок: ',*list(directories.keys()), sep='  ')    
    return
            
  
    
command=input('Введите команду: ')    
while command!='q':
    if command =='p':
        num1=input('Введите номер документа: ')
        nd=p(num1)
        if nd<len(documents):
            print( 'Владелец документа: ',documents[nd]['name'])
        else:
            print('Документ не найден в базе')
    elif command == 's':
        num1=input('Введите номер документа: ')
        shelf=s(num1)
        if shelf<=len(directories):
            print('Документ хранится на полке: ',shelf )
        else:
             print('Документ не найден в базе')
    elif command == 'l':    
        l()
    elif command == 'as': 
        owner=input('Введите номер полки: ')
        a_s(owner)
    elif command == 'ds':    
        owner=input('Введите номер полки: ')
        ds(owner)
    else:
        print('Нет такой команды')
    command=input('Введите команду: ')
print('Программа завершена')
```
