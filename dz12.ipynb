```python
import pandas as pd
```

```python
data = pd.read_csv('ratings.csv')

def rating(param):
    if param <= 2:
        return('Низкий рейтинг')
    elif param <= 4:
        return('Средний рейтинг')
    else:
        return('Высокий рейтинг')
    
    
    
data['class'] = data['rating'].apply(rating)   
data.head(30)
```

```python
data = pd.read_csv('keywords.csv')
data.head(70)
geo_data = {

'Центр': ['москв', 'тул', 'ярославл'],

'Северо-Запад': ['петербург', 'псков', 'мурманск'],

'Дальний Восток': ['владивосток', 'сахалин', 'хабаровск']

}

def region(param):   
    for i in geo_data:        
        for j in geo_data[i]:
                if j in param :
                      return(i)
           
    return('undefined')  
        
data['region'] = data['keyword'].apply(region)  
data.head(130)
```
